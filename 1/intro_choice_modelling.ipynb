# Introduction to Choice Modelling



## What is Choice Modelling?

Choice modelling or 'Discrete Choice Modelling' is the statistical analysis of individuals' choices. What are the factors that influence a particular choice, such as the choice of transport method to get to work? Can we predict  what someone is going to choose if we know some variables about the individual and their choice alternatives? For example: What would be the choice of smartphone, knowing the brands, prices available and the individual's age group and income?

Perhaps unsurprisingly, choices are involved in many situations of everyday life, so the applications of Choice Modelling are very varied.
Examples:
 * Transportation.
 * Marketing.
 * Policy making.
 * Real State.
 * Hospitality (Choosing a restaurant).
 * Politics.
 * Insurance.
 * Criminology.


Choice modelling is rooted in Statistics and Economic Theory.
In a classical regression, we look for the relationship between **numerical** dependent variables and explanatory variables. But we rarelly observe choice in numerical form, often we can only observe the result of the choice (whether we drive to work, cycle or take the bus). To analyze this type of data, a new kind of mathematical framework had to be developed. The purpose of this unit
is to give the fundamentals of these mathematical methods so they can be applied in practice. Students will also be able to understand research studies based on these techniques.

## 什么是选择建模？

选择建模或“离散选择建模”是对个体选择的统计分析。影响特定选择的因素是什么，例如选择通勤工作的交通方式？如果我们了解个体和他们的选择备选项的一些变量，我们能预测他们会选择什么吗？例如：如果知道品牌、价格和个体的年龄组和收入，能预测他们会选择哪种智能手机？

或许并不奇怪，选择涉及日常生活中的许多情况，因此选择建模的应用非常广泛。
例如：
* 交通运输。
* 市场营销。
* 政策制定。
* 房地产。
* 餐饮业（选择餐厅）。
* 政治。
* 保险。
* 犯罪学。

选择建模根植于统计学和经济学理论。
在经典回归中，我们寻找**数值**因变量与解释变量之间的关系。但我们很少以数值形式观察选择，通常我们只能观察选择结果（例如是否开车上班、骑自行车还是乘公交车）。为了分析这类数据，必须开发一种新的数学框架。本课程的目的是介绍这些数学方法的基本原理，以便能够在实践中应用。学生还将能够理解基于这些技术的研究成果。

## The BART case

 In 1971, a new electric railway system was being constructed around the San Francisco area in the US, the BART. Daniel McFadden and his team had been working on ideas that would later become know as 'discrete choice modelling'. They set up to forecast the demand for this system before it began operation to prove the capacity of the methods they were developing.

**A bit of context:** To get some perspective, it helps to get a notion of how this problem, predicting demand for a new transport system,
was being approached *before* choice modelling was popularized, so we can compare choice modelling with other methods. The method was the so-called gravity model, a method working at the aggregate level (areas, populations), rather than the individuals (persons).

The gravity method can be roughly described as follows:


每次我们上班时，我们都需要做一个关于如何去上班的选择：
我们应该开自己的车吗？拼车？还是乘公交车？

交通运输是离散选择最受欢迎的应用之一，而湾区快速交通（BART）的案例研究可能是离散选择建模应用的最著名示例，并且是其潜力的一个优秀说明。历史意义：这个项目的研究者丹尼尔·麦克法登于2000年获得诺贝尔经济学奖，以表彰他对离散选择理论和方法的发展。

## BART案例

1971年，美国旧金山地区正在建设一种新的电动铁路系统，即湾区快速交通（BART）。丹尼尔·麦克法登及其团队一直在研究后来被称为“离散选择建模”的想法。他们着手在该系统开始运营之前预测对该系统的需求，以证明他们正在开发的方法的可行性。

**一些背景信息：** 为了从某种程度上了解在离散选择建模普及之前，对新交通系统需求进行预测是如何被处理的，我们可以将离散选择建模与其他方法进行比较。那时使用的方法被称为重力模型，该方法在聚合水平（地区，人口）而非个体（个人）上进行工作。

重力模型可以粗略地描述如下：


![gravity from http://evertmeijers.com/new-article-presents-novel-method-to-identify-intercity-relationships/networks-nl-gravity-model/](https://raw.githubusercontent.com/pmontman/pub-choicemodels/main/img/gravity_example_nl.png "taken from http://evertmeijers.com/new-article-presents-novel-method-to-identify-intercity-relationships/networks-nl-gravity-model/")



1. Divide the urban area into small zones.
2. Measure how often people travel and the origin and destination of the trips. This can be done through a survey of people living in these zones.
3. Aggregate count to determine the 'flow' between each pair of travel zones.
4. These flows are proportional to the 'size' or 'mass' of each zone, measured as the number of people or jobs they contain. The flows are *inversely* proportional to the travel time between zones or a more sophisticated measure that considers 'travel cost' (price of ticket, fuel, time).

The result is something similar to the law of gravitational attraction.

1. 将城市区域划分为小区域。
2. 测量人们的出行频率以及出行的起点和目的地。可以通过对居住在这些区域的人进行调查来实现。
3. 汇总计算，确定每对出行区域之间的“流量”。
4. 这些流量与每个区域的“大小”或“质量”成比例，可用人数或包含的工作岗位数量来衡量。这些流量与区域之间的出行时间成*反比*，或者使用更复杂的衡量方式来考虑“出行成本”（票价、燃料、时间等）。

结果类似于万有引力定律。


![gravity masses](https://raw.githubusercontent.com/pmontman/pub-choicemodels/main/img/gravity_masses_plot.jpg "gravity masses from https://newellta.weebly.com/gravity-model.html")



To use the gravity model for planning, we can introduce the impact of the planned changes either in 'distance' (travel cost) (new bus routes, increased highway capacity, pricing adjustments) or 'mass' (new office buildings, new or growing residential zones) and calculate the resulting flows for this hypothetical scenario.

Gravity models are useful tools, still being applied, but McFadden postulated that they had two major limitations.
1. By working on aggregates, they lose important fine-grain information than might be used for prediction.
2. Some transportation policies are difficult to include in the model, such as a completely new railway system.

McFadden's idea was to predict travel demand at the individual level (each person) by collecting data at a sufficienly detailed level: distance to a bus stop, waiting times, socioeconomic attributes of the individuals (bus route to a wealthy neighbourhood), etc. If we then predict the choices for the individuals under the expected new conditions (BERT is introduced) and aggregate these choices to the population level (count how many times each option is predicted to be chosen), we can get a better approximation of what will be the demand for the different modes of transportation.
 The slogan of the project was:
> "*Zones don't travel; people travel!*"


使用重力模型进行规划时，我们可以引入计划变化的影响，无论是在“距离”（出行成本）方面（如新的公交线路、增加高速公路容量、价格调整），还是在“质量”（新办公楼、新的或不断扩大的居住区）方面，并计算出该假设场景下的结果流量。

重力模型是有用的工具，仍在应用中，但麦克法登提出了它们有两个主要局限性。
1. 通过对聚合数据进行处理，他们丢失了可能用于预测的重要细粒度信息。
2. 有些交通政策难以纳入模型中，比如完全新的铁路系统。

麦克法登的想法是通过在足够详细的层次收集数据（例如距离公交车站的距离、等待时间、个人的社会经济属性（公交车路线通往富裕社区）等）来预测个体级别（每个人）的出行需求。然后，如果我们根据预期的新条件（引入BERT）预测个体的选择，并将这些选择聚合到人口层面（计算每个选项预计被选择的次数），我们可以更好地近似不同交通方式的需求。该项目的口号是：
> “*地区不旅行；人们旅行！*”


## The choice model for BART

Lets start by the results.
During 1972, *before the railway system was in operation*, the team gathered information about individuals in a large survey and the *predicted* the adoption of the new railway system using a choice model based on that data. Years later, when the service have been operative for some time, they *reinterviewed* the same subjects of the survey and compared their original predictions to the actual adoption.

The official predictions of adoption for BART, based on a 'optimistic' version of a gravity model, were of 15% of all work trips. The prediction of the choice model was 6.3%. In 1975, the actual adoption was **6.5%**. Thus the predictions of the choice model were remarkably accurate. At the moment we do not know much about the specifics of the choice model, but we can think for the time being that we are working with a linear regression model *(as a very crude approximation!)*. We fit a 'linear regression model' to each choice alternative.

So the variables are measured for each combination of 'transport method' and 'individual'. These variables are multiplied by some numbers (the *coefficients* of the model) and then everything is added together to compute a final number. The larger this final number, the more likely that that transport method is chosen by that individual. Without getting into much detail, we can take a look at the sign of the coefficients (the 'Max. For example, the variable 'Number of transfers', measured for people that travel by bus, has a negative coefficients (-0.1050), meaning that the more transfers an individual has to make to get to work by bus, the less likely they will be to choose to go by bus. Other variables have interesting economic interpretation, for example, time spent either walking, in a car or on the bus each have a different coefficient (all negative as expected) but they can be compared. Looking at the coefficients and comparing to the wages, McFadden estimates the 'value of time as a percent of the wage' for each of these methods of transportation. For example, the value of time spent in a car in 226% the value of the wages, meaning that time spent in commuting by car is more grueling than work. In contrast, value of time bus time is 91%, commuting by bus is more confortable than by car. These are the kinds of intuitions that we can get from choice models.

这些变量以及其他一些变量被引入到模型中，并估计它们的影响，得到下表。目前我们对选择模型的具体细节了解不多，但可以暂时认为我们在使用一个线性回归模型（非常粗略的近似！）。我们将“线性回归模型”适用于每个选择备选项。

因此，这些变量为“交通方式”和“个体”的每种组合进行测量。这些变量与模型的一些数字（模型的系数）相乘，然后将所有结果相加以计算出最终数字。这个最终数字越大，该个体选择该交通方式的可能性就越大。不深入细节，我们可以查看系数的符号（“最大似然估计”列），以确定增加某个变量值对采用特定方式的可能性是否产生正面或负面影响。

例如，针对乘坐公交车的人测量的“换乘次数”变量具有负系数（-0.1050），这意味着一个人通勤时需要进行的换乘次数越多，他们选择乘坐公交车的可能性就越小。其他变量具有有趣的经济解释，例如，步行、开车或乘坐公交车所花费的时间都有不同的系数（都为负值，符合预期），但它们可以相互比较。通过查看这些系数并与工资进行比较，麦克法登估计了每种交通方式的“时间价值与工资的百分比”。例如，开车花费的时间价值相当于工资的226%，意味着开车上下班所花费的时间比工作更加辛苦。相比之下，乘坐公交车花费的时间价值为91%，说明乘坐公交车上下班比开车更舒适。这就是我们从选择模型中可以得出的一些直觉。

## Introducing the railway system into the model

Notice that the model does not include BART as a method of transportation.
This is not suprising, since BART was not operative when the survey was conducted, and the actual purpose of the project was to predict was going to happen with BART when it started running!

How can we use the model to predict the share of BART?
After adjusting for the 1975 travel time and costs (remember the survey was conducted in 1972) the team considered a new model with the BART travel alternatives, with coeffcients set to be equal to the bus. So the value of time, effect of the number of conmutes, walking distance to a stop, etc. for the railway system was assumed by the expert to be the same as for the bus. What changed are the actual values of the variables (e.g. less travel time for BART than for bus for the same trip). This is an important illustration on the use of choice model to predict the adoption of a new choice never seen before.

## 引入铁路系统到模型中

请注意，模型并不包括BART作为一种交通方式。这并不奇怪，因为在进行调查时，BART尚未开始运营，而该项目的实际目的是预测BART开始运营后会发生什么！

我们如何使用模型来预测BART的份额？
在调整了1975年的出行时间和成本（请记住，调查是在1972年进行的）后，团队考虑了一个包含BART出行选择的新模型，并将其系数设置为与公交车相等。因此，对于铁路系统，时间价值、通勤次数的影响、步行到车站的距离等与公交车被认为是相同的。变化的是实际的变量值（例如，BART与公交车相比，同一次行程所花费的时间更少）。这是使用选择模型来预测新的从未见过的选择采用情况的重要说明。

Another historically relevant application of choice modelling, this time
a [major study in marketing](https://doi.org/10.1287/mksc.1070.0331). This was in published in 1982 with
the experiment ending (the data gathering) in 1980.
Using supermarket scanner data, a new technology at that time, the researchers
studied the driving factors of household's choice for coffee brand and package sizes. By appling choice models, they inferred how sales are influenced by aspects such as brand loyalty, package size loyalty, presence of promotions, price, promotional price cuts.

An especially interesting aspect of the model is that it goes beyond the
variables affect each invidivual product's sales, it provides information about interdepencies of products within the same category. For example, a price cut might increase the sales of one product at the expense of other of a different brand, or from another size packaging of the same brand. These types of information are useful for manufacturers.

## Experiment details

The data gather was done on four supermakets in Kansas City, collected from September 1978 to March 1980. They tracked the purchases of 2000 households (with their consent),
including products bought, date of the purchase and price. After studying the quality of the data, the researchers settled for studying the purchase patterns of only ground coffee, across 5 different brands and two package sizes (small and large).

The households were divided into two groups, one was used to estimate the model and the other to test the quality of the estimated model.
Furthermore, the tracking periods were also divided in two, the first 32 weeks used to estimate the model, the last 20 weeks to test how the model does when predicting. This experimental setup therefore allows to test how well the model predicts the purchasing patterns:
 * New, unseen households.
 * 'Into the future', unseen time periods.

**Which variables were chosen?**

 * Interestingly, because scanner data allowed tracking purchases of each household across time, one of the variables of the model was the number of previous purchases within the same brand for a given household: 'brand loyalty'.
 * Similarly, the concept for 'size loyalty' was measured, the number of purchases with the categories of small packages or large packages.
 * Price of the product at the moment being bought.
 * If there is advertisement in the newspaper.
 * In order to track the effectiveness of promotions, whether they are effective in making the household purchase more of that brand afterwards, they measured if previous purchases of that household were of a promotion within the same brand. This can measure if the promotion make customer repurchase that brand when it is no longer in promotion, 'switching'.
 * Note that no socioeconomical variables from the household (income, family structure..) were tracked, the 'loyalty' or 'promotional switching' variables could hypothetically account for that.

This modelling approach is very simple compared to the previous transportation example, measuring way fewer variables.

## Results

The first result to highlight is that the model predicts remarkably well.
Below, a plot of the predictions vs the actual shares of a product (small packaging of brand A). The shares are predicted for the houldout set of families, the one that were not used for estimating the model.
the share are tracked across time, week by week. The vertical line marks the time period that was not used for estimated the model, dates after that line representing the 'future'.
 * The model predicts well for new individuals, unlike the BART case which is applied to the same individuals.
 * The model predicts well on different supermarket types (large, neighbourhood, warehouse,...), even when superkmarket type is not considered a factor in the model.


另一个在历史上相关的选择建模应用，这次是[市场营销的主要研究](https://doi.org/10.1287/mksc.1070.0331)。该研究于1982年发表，实验数据采集在1980年结束。
研究人员利用当时的新技术——超市扫描仪数据，研究了家庭在咖啡品牌和包装尺寸选择方面的驱动因素。通过应用选择模型，他们推断了销售受到品牌忠诚度、包装尺寸忠诚度、促销活动的影响，以及价格、促销价格折扣等方面的影响。

模型的一个特别有趣的方面是它不仅考虑了影响每个个体产品销售的变量，还提供了有关同一类别内产品之间相互依赖关系的信息。例如，降价可能会增加某个产品的销售量，但可能会损害不同品牌的其他产品或者同一品牌的不同包装规格的产品。这些信息对制造商非常有用。

## 实验细节

数据采集在堪萨斯城的四家超市，时间跨度为1978年9月至1980年3月。研究人员跟踪了2000个家庭（在得到他们的同意后）的购买情况，包括购买的产品、购买日期和价格。在研究数据质量后，研究人员决定仅研究咖啡的购买模式，涵盖5个不同品牌和两种包装规格（小包装和大包装）。

家庭被分为两组，一组用于估计模型，另一组用于测试估计模型的质量。
此外，跟踪期间也被分为两个部分，前32周用于估计模型，后20周用于测试模型在预测时的表现。因此，这个实验设置允许测试模型在以下情况下的预测情况：
* 新的、未知的家庭。
* '未来'，未知的时间段。

**选择了哪些变量？**

* 有趣的是，由于扫描仪数据允许跟踪每个家庭的购买历史，模型的一个变量是给定家庭在同一品牌下的先前购买次数：“品牌忠诚度”。
* 同样地，也测量了“尺寸忠诚度”的概念，即购买小包装或大包装的次数。
* 购买时产品的价格。
* 报纸上是否有广告。
* 为了追踪促销活动的效果，他们测量了家庭的先前购买是否属于同一品牌的促销。这可以衡量促销活动是否使顾客在促销结束后继续购买该品牌，即“转换”。 
* 需要注意的是，没有跟踪家庭的社会经济变量（收入、家庭结构等），这些“忠诚度”或“促销转换”的变量可以在假设上代表这些因素。

与前面的交通案例相比，这种建模方法非常简单，测量的变量要少得多。

## 结果

值得强调的第一个结果是，模型的预测效果非常好。
下图显示了一个产品（品牌A的小包装）的预测份额与实际份额的对比。这些份额是为没有用于估计模型的待选家庭预测的。份额随时间变化，逐周追踪。竖线标记着未用于估计模型的时间段，该线之后的日期代表“未来”。
* 与应用于相同个体的BART案例不同，模型对新个体的预测效果良好。
* 模型在不同类型的超市（大型、社区、仓库等）上的预测效果也很好，即使超市类型在模型中并没有被考虑为一个因素。


![coffee prediction](https://raw.githubusercontent.com/pmontman/pub-choicemodels/main/img/choicecoffee_prediction.PNG)




The outstanding accuracy of the predictions justifies further analysis of the model, for example, analyzing the coefficients. IF the models were not very accurate, interpretation should be taken with a grain of salt. Even when models are accurate we should be careful, we should replicate the experiment before drawing strong conclusions, because the model was designed after the data was being gathered.

 * they find that tracking loyalty improves the model, one of the more important attributes to measure.
  * While brand loyalty is important, 'size loyalty' is also very important, which is unexpected in the view of the researchers, arguing that this is new relevant information for manufacturers and retailers.
 * Again authors attribute the success of the experiment to the individualized analysis, the individuals being households. The market shares are deduced from individual choices.
 * The model can be used to estimate market response to advertisement,  price elasticity and price cuts elasticity, both normal and under promotion.
 * They identified two groups of customers the 'loyal customers' not affected by marketing actions and the 'switchers' which are affected.


## A take away
Barcode scanning was a new technology back then, and this experiment delivers a powerful way of taking advantage of an unprecedented quality and quantity of data. We now live in a time were new sources of data are constantly appearing, it is useful to have the tools to make the most of them.

* 他们发现跟踪忠诚度可以提高模型的效果，这是一个更重要的要素需要衡量。
* 尽管品牌忠诚度很重要，但“尺寸忠诚度”也非常重要，这是研究人员意料之外的发现，并认为这是对制造商和零售商的新有价值信息。

This time we have an example of a 2011 [choice experiment](https://onlinelibrary.wiley.com/doi/10.1111/j.1467-8489.2011.00551.x) in environmental policy.
Choice modelling is used to understand how Australian citizens value the 'environmental quality' of the area around the Murray river, of great agricultural, environmental and cultural importance in Australia. The results of the experiment would help decision-makers allocate resources to stop or revert the decline of that area.

## The questionnaire

The researchers surveyed a sample of over 3000 Australian households.
Participants received in their mail a questionnaire and some side information on the River Murray.

 * They were told that the quality of the River Murray had been declining for the past 40 years.
* Causes of the decline were described, such droguhts, irrigation of neighbouring zones, non-native species and so on.
 * The changes in quality were described in 4 categories:
   * Waterbird breeding along the River Murray
   * Native fish populations in the River Murray
   * Healthy vegetation along the River Murray
   * Waterbird habitat in the Coorong
 * Information of the ways to improve quality were described, including:
   * The government purchasing water from farmers willing to sell in water markets.
   * The government providing incentives to increase irrigation efficiency, habitat rehabilitation and engineering works.
  * They were told the cost of each of the changes per household, resulting in a 10-year period of increased taxes and food costs.



Unlike previous experiments, this time the researchers created the set of options that was presented to the individual choice-makers. The researchers carefully *designed* the options that the subjects had to choose from. Not all household received the same set of options, they were not asked the same questions.
The reason is that the amount of possible options in the case of policy making is overwhelming, so it is 'impossible' to ask households to choose between them. Imagine all the possible combinations on the actions that can be taken along with their costs in all of the 4 categories, and you are asked to choose which one you prefer. It can be quite cumbsersom to make a decision.

*When the choice is large and difficult, participants will choose to opt out of the experiment (they will not answer) or become tired and answer 'randomly'.*

Another important difference with previous case studies: in this experiment the options are artificial, they are not present in the environment, such as choosing transport method or a brand of coffee.

 The participants received 4 questionnaires similar to the one presented in the Figure below, they had to choose among 3 alternatives. **So they are presented a choice among 3 alternatives**. They have to repeat the process 4 different times (already quite tedious), each time the set of alternatives being slightly different.
With this design, the researchers could approximate what would be the result of asking the participants to choose among the 'real' underlying large set of alternatives. The experimental design is 'efficient' in the sense that the 'real thing' can be approximated without having to ask the participants to choose among many alternatives.
 **We will learn about the design of these types of experiments and how to make them efficient.**

The researchers measured attributes of the alternatives
 * cost.
 * increase in quality.
 * increase in population of native fish.
 * increase in frequency of waterbird breeding.
Along with socieconomic variables of the respondents
 * gender.
 * age.
 * income.
 * region they live in.
 * number of children.

 这次我们有一个涉及环境政策的[2011年选择实验](https://onlinelibrary.wiley.com/doi/10.1111/j.1467-8489.2011.00551.x)的例子。
选择建模被用来了解澳大利亚公民对澳大利亚重要的农业、环境和文化价值的穆雷河周围地区的“环境质量”有多重视。实验结果将帮助决策者分配资源，以阻止或逆转该地区的衰退。

## 问卷调查

研究人员对3000多个澳大利亚家庭进行了样本调查。
参与者在邮件中收到一份问卷调查和关于穆雷河的一些附加信息。

* 他们被告知穆雷河的质量在过去40年里一直在下降。
* 下降的原因被描述了出来，例如干旱、邻近地区的灌溉、非本土物种等。
* 质量变化被描述在4个类别中：
  * 穆雷河的水鸟繁殖情况
  * 穆雷河的本土鱼类数量
  * 穆雷河沿岸的健康植被
  * 鸬鹚栖息地在库朗地区
* 还描述了改善质量的方法，包括：
  * 政府从愿意在水市场出售的农民那里购买水。
  * 政府提供激励措施，以提高灌溉效率、恢复栖息地和进行工程工作。
* 他们被告知每个改变的成本，每户家庭将在10年内面临增加的税收和食品成本。

与以前的实验不同，这次研究人员创建了提供给个体选择者的选项集。研究人员精心“设计”了被调查对象必须选择的选项。并不是所有家庭都收到相同的选项集，他们没有被问到相同的问题。
原因是在政策制定的情况下，可能的选项数量是庞大的，所以“不可能”要求家庭在其中进行选择。想象一下所有可能的行动组合以及所有4个类别中的成本，然后要求你选择你更喜欢哪一个。这可能会非常繁琐。

*当选择范围较大且难以做出决策时，参与者会选择退出实验（不会回答问题）或变得疲倦并“随机”回答问题。*

与之前案例研究的另一个重要区别是：在这个实验中，选项是人为构建的，它们并不存在于环境中，比如选择交通方式或咖啡品牌。

参与者收到类似下图所示的4份问卷调查，他们必须在3个选择中进行选择。**所以他们面临的是在3个选择中进行选择**。他们必须重复这个过程4次（已经相当繁琐了），每次选择的选项集都略有不同。
通过这种设计，研究人员可以近似地了解参与者在“真实”的庞大选项集中进行选择时的结果。实验设计在“有效”意义上是高效的，因为无需要求参与者在众多选项中进行选择即可近似“真实”的情况。**我们将学习关于这类实验的设计以及如何使其高效。**

研究人员测量了选项的属性
* 成本。
* 提高质量。
* 增加本土鱼类数量。
* 增加水鸟繁殖频率。
以及调查对象的社会经济变量
* 性别。
* 年龄。
* 收入。
* 居住地区。
* 孩子数量。

## Results

The researchers were able to estimate the 'willingness to pay' for each the changes, for example:
 * Increase the native fish population by 1% each year per 10 years was estimated between \$1.71 to \$3.58 per year.
 * Increase the health vegetation along the river by 1% each year was estimated between \$2.87 to \$4.42 per year.

The researchers found that different regions (Victoria, NSW, ...) had different willingness to pay values.

Overall, the findings  "indicate that respondents are willing to pay substantial amounts to increase environmental flows and improve the quality of the Murray River and Coorong". The reserarches estimated a "willingness to pay of $13 billion for good-quality waterbird habitat in the Coorong and a medium to upper end set of improvements along the River Murray." Again the choice model studies the individual and then the results are aggregated up for estimates at the population level, in this cases the states that are crossed by the Murray river.

An important difference with previous studies: this study measures hypothetical choices, not real ones.

## 结果

研究人员能够估计每项改变的“愿意支付”金额，例如：
* 每年增加本土鱼类数量1％，估计每年为10年，介于1.71美元至3.58美元之间。
* 每年增加沿着河流的健康植被1％，估计每年为10年，介于2.87美元至4.42美元之间。

研究人员发现不同地区（维多利亚州，新南威尔士州等）具有不同的愿意支付价值。

总体而言，研究结果“表明受访者愿意支付相当大的金额来增加穆雷河和库朗地区的环境流量和改善水质”。研究人员估计“愿意支付130亿美元用于库朗地区优质的水鸟栖息地，以及一系列中到上等级的穆雷河改善措施。” 再次强调，选择模型研究个体，然后将结果汇总得出人口层面的估计，这种情况下是跨越穆雷河的州。

与之前的研究相比，这项研究的一个重要区别是：这项研究测量的是假设性选择，而不是实际选择。

#Example 4) Smartphone choice in South Korea

A [2019 study on smartphone preference is South Korea.](https://doi.org/10.1016/j.jretconser.2019.101907) Researchers want to understand what drives the interest in the different brands of smarphones and the different characteristics of their models (screen size, cpu, OS, memory, face recoginition, price). 4 smartphone brands are studied: Apple, Samsung, LG and others.

Some highlights of their study (quoting):
* Analyze the consumers’ brand loyalty and interest for smartphones in South Korea.
* Demonstrate Samsung has the highest interest and Apple has the highest loyalty.
* Examine the effect of innovative peers on brand loyalty and interest.
* Find that peers with high innovativeness affect only to Apple's brand loyalty.

"In order to analyze the actual effect of brand on consumers’ preference, the discrete choice model should be applied. It is
because the discrete choice model based on random utility framework
allows for evaluating the trade-off between attributes (Folta, 1998). The
discrete choice model has an advantage that can provide the results
such as willingness to pay and relative importance that can be useful to
implement in marketing strategies (Lee et al., 2006)."



## The experiment

Researches surveyed a sample of 991 citizens living in Seoul and other South Korean cities, asking them to choose an smartphone among a set of alternatives.

The large range of smartphones presented a problem similar to the environmental case study, they had 384 different types of smartphone when considering attributes such as
 * Brand
 * Screen size
 * Prize
 * Memory
 * Presence of user recognition techonology (face recognition).

To prevent decision fatigue from the subjects, an special experimental design
was used. The subjects were presented a limited
sets of alternatives to choose from, just 4 alternatives.

Socieconomic characteristics of the subjects were also considered:
 * Age
 * Gender
 * Region
 * Income

Of special interest in this study are two additional characteristics:
 * A 'reference point': the phone that they are already using.
 * 'Peer effect': Subjects were asked to nominate a person they know, whom they rate as 'innovative' (the friend that tries new technologies). They were asked to rate their friend satisfaction on their brand of smartphone.

## 实验

研究人员对生活在首尔和其他韩国城市的991名市民进行了调查，要求他们在一组备选项中选择一款智能手机。

众多智能手机的选择对于决策疲劳的问题，与环境案例研究类似，当考虑诸如以下属性时，他们有384种不同类型的智能手机：
* 品牌
* 屏幕大小
* 价格
* 内存
* 是否有用户识别技术（面部识别）。

为了防止被调查对象的决策疲劳，采用了特殊的实验设计。被调查对象只会被展示一个有限的备选项集，仅包含4个备选项。

被调查对象的社会经济特征也被考虑在内：
* 年龄
* 性别
* 地区
* 收入

这项研究中特别感兴趣的是另外两个特征：
* “参考点”：他们已经在使用的手机。
* “同伴效应”：被调查对象被要求提名他们认识的、他们认为“创新”的人（尝试新技术的朋友）。他们被要求评价他们的朋友对所使用的智能手机品牌的满意度。


## Reference point

An important point in this study is the focus on a **reference point** for each consumer: the phone that they are already using and how this affects their smartphone preference. They think that some factors such as having to change the OS has influence on the preferences. The importance of using a reference point, which somewhat 'challenges' basic economic theory, is postulated by Daniel Kahneman, another Nobel prize in Economic Sciences (Year 2002). The book 'Thinking, Fast and Slow' is an interesting read.

 We have already seen the idea of reference point studied in the Coffee - Barcode scanning example, where the researchers tracked previous purchases of the household as a characteristic in their model.

Quoting the paper: "In general, consumer preference is accepted to be unrelated to the reference point or the status quo in standard consumer choice models
concerning consumers' purchase decisions, including the discrete choice
model." "...In addition, in the
case of South Korea, where the majority of consumers already use
smartphones, applying a reference point to the model would yield more
meaningful results (Kim et al., 2016; Kim et al., 2018)."

## 参考点

这项研究的一个重要点是每位消费者的**参考点**：即他们已经在使用的手机以及它如何影响他们对智能手机的偏好。他们认为一些因素，比如需要更换操作系统，会影响到偏好。使用参考点的重要性在经济学领域也被丹尼尔·卡尼曼提出，他是另一位获得诺贝尔经济学奖的学者（2002年）。他的著作《思考，快与慢》是一本有趣的阅读。

我们已经在咖啡-条形码扫描的例子中见过参考点的概念，研究人员将家庭的之前购买情况跟踪作为模型中的一个特征。

引用论文中的话：“通常，消费者偏好被认为与参考点或现状在标准的消费者选择模型中无关，涉及消费者的购买决策，包括离散选择模型。”“...此外，在韩国这种大多数消费者已经使用智能手机的情况下，将参考点应用于模型将产生更有意义的结果（Kim等，2016；Kim等，2018）。”

# 分析

研究人员对数据应用了三种不同的选择模型进行分析。在每个分析中，他们都添加了一个相关的特征到之前的模型中。这样可以估计该特征对更简单实验的影响，而不是直接对数据进行分析。这种比较的方法从最简单的模型开始，逐步增加复杂性，帮助我们首先建立一个基准，然后衡量改进相对于基准的效果是否有用。

1. 只使用手机特性和用户的社会经济变量。
2. 使用手机特性、社会经济变量和用户当前拥有的手机作为参考点。
3. 完整实验考虑了手机特性、社会经济变量、参考点以及用户认为具有创新性的“同伴”的满意度。

让我们来看看第一个分析的结果，只考虑了手机特性和社会经济变量。模型的系数在下表的“Mean”列下。简单来说，品牌特征的数字越大，用户选择该品牌手机的可能性就越大。其他特征具有预期的符号，屏幕大小越大，用户选择该手机的可能性就越大，存储容量和面部识别也是如此。价格有负号，所以价格越高，用户选择该手机的可能性就越小。一个简单的解释是，如果所有其他特征保持不变（屏幕大小、存储容量、价格等等都相同），那么三星手机比其他品牌的手机更有可能被选择。这也是符合预期的，因为三星是韩国的热门品牌。

smartphone_phoneonly.PNG

Now lets consider the second experiment, where they included the 'reference point' in the characteristics they track.
We will see int the next table that all variables now have two coefficients 'Gains' and 'Losses'.
Gains represents the positive effect of changing from the reference point to a smartphone that has a different value of that attribute. 'Losses' would represent the influence of leaving that brand. For example 'Gains' on brand variable for 'Samsung' represents the increased preference for a Samsung phone of a user that does not have a Samsung at the moment. 'Losses' would be the aversion to change to another brand for a user that is using a Samsung at the moment.
We see that the 'Losses' have larger effects than 'Gains', indicating that users have an aversion to change smartphone brand.

For example, an user currently owning a Samsung and changing to Apple would derive a 'Gain' of 2.305 and a Loss of -6.632, indicating that they are unlikely to change, unles other characteristics such as prize, screen size and so on compensate for that.

 **The strongest loss appears in the Apple brand, indicating that Apple users are the most 'loyal'. The stronges gains appear in Samsung, which indicates the brand with the highest interest.**

Notice how much more nuanced is this analysis compared to the previous case without reference point.

现在让我们考虑第二个实验，在这个实验中，他们在所跟踪的特征中包含了“参考点”。
在下表中，我们将看到所有变量现在都有两个系数，“增益”和“损失”。
增益表示从参考点变换到具有不同属性值的智能手机的积极效果。“损失”表示离开该品牌的影响。例如，对于品牌变量“三星”，“增益”表示目前没有使用三星手机的用户对三星手机的偏好增加。“损失”表示正在使用三星手机的用户对切换到其他品牌的抵触。我们可以看到“损失”比“增益”有更大的影响，这表明用户对更换智能手机品牌有抵触情绪。

例如，一个目前拥有三星手机并转换到苹果手机的用户，将获得2.305的“增益”和-6.632的“损失”，这表示他们不太可能更换手机，除非其他特征，如价格、屏幕大小等可以弥补这一点。

**最强的损失出现在苹果品牌，这表明苹果用户最“忠诚”。最强的增益出现在三星，这表明三星是最受欢迎的品牌。**

请注意，与之前没有参考点的情况相比，这种分析要复杂得多。

The final analysis includes the effect on an 'innovative peer'. The results are more difficult to read, but the researchers highlight.

 * The inclusion of peer effect in the model does not change the relative effects of interest and loyalties among brands (Samsung still most interesting, Apple most loyal.
 * About the peer effect: "..consumers had a higher interest in brands that they do not own when their peers with a high level of innovativeness expressed satisfaction with that brand of
smartphone. However, brand satisfaction felt by peers with higher innovativeness was found to be a statistically significant influencer of
brand loyalty only for Apple."


 **This is an example on why it is important to check the predictive capability of the model, because you might interpret a model and draw conclusions, but then the model does not predict well or there is an alternative model that does better, how meaningful are the conclusions?** In this case the researchers do not check predictive capabilities, though they apply some statistical analysis to check which of the models should predict better.

最终的分析包括对“创新同伴”的影响。结果更难读懂，但研究人员强调了以下几点：

* 在模型中加入同伴效应不会改变品牌间兴趣和忠诚度的相对影响（三星仍然是最有吸引力的，苹果最忠诚）。
* 关于同伴效应：“...当消费者的高度创新型的同伴对某个品牌的智能手机表示满意时，他们对不拥有该品牌的兴趣会更高。然而，同伴的品牌满意度对于苹果品牌的忠诚度影响是具有统计学意义的。”

**这是一个例子，说明检查模型的预测能力是很重要的，因为你可能对一个模型进行解释并得出结论，但是后来发现该模型的预测效果并不好，或者有另一个模型效果更好，那么之前的结论有多有意义呢？** 在这种情况下，研究人员并没有检查预测能力，虽然他们对哪个模型应该预测得更好进行了一些统计分析。
