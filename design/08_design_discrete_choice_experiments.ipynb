# The goal of Design of Experiments

To discover the underlying 'model' with as much efficiency as possible.
This means getting the most accurate estimation of the parameters in the model at a given 'budget' of sample size.

It turns out that this is a nontrivial pursuit, basically we can use 'Statistics' to design better experiments!

Design of experiments implies much more agency from the analyst in the gathering of the data, compared to trying to come up with a good predicive model given a dataset.

Imagine that we want to design a product (a smartphone) or a 'menu' (our offer for coffe types). What variables to measure, how many different values? How many individuals we should sample? How do we present the choices to them?

# Terminology

An **experiment** defined in scientific terms involves the observation of the effect upon one variable, a response variable, given the manipulation of the levels of one or more other variables.

The design of experiments is a field that has developed its own terminology, so there are again serveral ways of naming the same components of an experiment:

 * **Factor** is an Attribute (price, screen size, travel time)
 * **Factor levels** are Attribute levels (different prices, sizes of the product).
 * **Treatment** The invididual level of each Attribute (size=4inches).
 * **Treatment combination**  What we call the alternatives (size=4inches, price=300 dollars)
 * **Choice sets or Profiles** Are the group of alternatives that are presented to the individuals for deciding between them.


 The manipulation of the levels of the variables does not occur in a haphazard manner. Rather, we turn to a specialized form of statistics to determine what manipulations to make, and when to make them. Thus we can say that the manipulations occur by design. Hence the name experimental design!

# Stated Preference and Revealed Preferences

* **Revealed Preferences**, we actually observe what individuals choose in a real choice situation.

* **Stated Preferences**, The response come from a hypothetical situation in a survey: "*Given this situation, what would you choose?*

The limitation of stated prefence data is, as we expect, there can be a huge difference between what you answer in a survey vs. what you really do. On the other hand, getting revealed preferences is many times impossible, for example, when doing a choice experiment for designing a product, we want to do the experiment before the product actually exists! So we can only deal with hypotheticals.

Ideally, we would go for revealed preferences data, but here are some arguments to motivate stated preference experiments:
 * Gather information about alernatives that do not exist (E.g. new products).
 * In real situations, some levels of the attributes do not actually exist (e.g. little variability in pricing of a product in a real market, experimeting with prices in a real situation is risky)
 * Attributes are correlated in real situations: E.g. size of screen of a smartphone and price are often correlated, so estimating the individual effects will be difficult (multicolinearity). Another example is cost and travel time in a transportation setting.
 * Stated Preference experiments are often much cheaper. *This summarizes all other points, if we had infinite budget, we could solve all other issues of revealed preferences.*

# Main effects and interactions

We define an effect as the impact a particular treatment has on some response variable. In choice analysis, the response variable is choice, the alternative selected.
Thus, an effect is the impact that a particular attribute level has on choice. We can think in terms of utility or choice probabilities. Choice probabilities complicates the processing, because it it nonlinear.

* **Main effect** is defined as the direct independent effect of each attribute upon the response variable, choice. The main effect, therefore, is the difference in the means of each level of an attribute and the overall or grand mean. E.g. price of 100 dollars decreases the utility by 20 utils, compared to the average utility for all observations in the dataset.
*  **Interaction effect** is an effect upon a response variable, choice, obtained by combining two or more attributes which would not have been observed had each of the attributes been estimated separately.

$$ V = \beta_1 x_1 + \beta_2 x_2 + \beta_{1,2}(x_1x_2)$$

Even though we have not seen many examples of interaction in other lectures, interactions become relevant in survey data becasue we have few variables to choose from: we the experimenters have to decide
the levels of the attributes, usually have to use just few levels to get the experiment 'manageable' in practice, so interactions can explain more of the unobserved utility compared to other scenarios where we have a lot of variables meausred from the alternatives.

# Types of encoding the effect
How do we encode the attributes in our model?

* **Linear:** The traditional 'values of the parameter' are multiplied by a number. In experimental design, the benefit is that we reduce the number of parameters to estimate in the model, compared to 'dummy' encoding.

* **Dummy:** (Non-linear, piecewise constant): Each level of the attribute will have produce a parameter to estimate. We have seen this type of encoding before, we can turn a numeric value into ranges, and then the ranges are encoded as dummies and we get a nonlinear especification with respect to the original untransformed variable. Because of the nature experiment design, dummy encoding is much more popular compared to applying the transfomation to field data. It also captures qualitave attributes (e.g. color of the smarthphone).

If we know that we can get away with a linear encoding, this can influence the decision on the 'levels' that we put in our experiment. For example, we can make the more separated than usual to get a better estimation of the cofficient. Example: Pricing range, instead of going for the a reasonable pricing of the alternatives, to better capture the influence of price, we can choose a bit more extreme pricing.

# Principles of Design of Experiments



Huber and Zwerina (1996) discuss four principles that impact the efficiency of
choice designs: level balance, orthogonality, minimal level overlap and utility balance.

Level balance and orthogonality are important concepts retained from linear design theory, whereas minimal level overlap and utility balanced are measures of the within choice set structure of the design. Satisfying all four of these criteria simultaneously is the most desirable condition, but near impossible to achieve for most practical applications.

 * **Level balance:** also referred to as balance (Payne 1988, Ball 1997) is the
requirement that all attribute levels occurs with equal frequency. For example, with a four level attribute, each level should occur in exactly one-fourth of the alternatives.

 * **Orthogonality:** Of main effects occurs when the "joint occurrence of any two levels of different attributes appear in profiles with relative frequencies equal to the product of their marginal relative frequencies (Addelman 1962)." In layman terms, this means that the level of the attributes and the interaction can be estimated without problems. In many practical applications level balance and orthogonality conflict and an improvement in one results in a degradation of
the other.

 * **Minimal level overlap**  is important in the structure of individual choice sets. Contrasts between attribute levels are only meaningful as differences within the choice set. Minimal level overlap means that the probability that an attribute level repeats itself in a choice set is as small as possible. The most extreme violation of minimal level overlap would occur when one factor has the same level across all alternatives within the choice set. Choices from this set provide no information about the participant’s
preferences for that particular factor.

 * **Utility balance** is achieved when the utilities of alternatives within choice sets are as close as possible. Achievement of utility balance requires some knowledge of the values of the parameters before construction of the design. Utility balance, along with the other three efficiency requirements is normally achieved and balanced through the maximization of Dp efficiency. *(This principle is actually in question, there are counterexamples...)*

# Fractional factorial designs

Fractional factorial designs are a common type of efficient design.
In a fractional factorial design, we are not interested in all interaction effects (they are not important, 0).
* This means that we do not need all possible combinations, they can be reduced.

* There are many fractional factorial designs that can be created from the full factorial reference. Which one we create depends on the prior information that we have, for example which interaction we know that will not have any influence.

* Nowadays, the way to look for the best factorial designs is more data-driven, this is, with some prior information (say, a pilot experiment) we find the factorial design tha maximizes a given criteria of efficiency.

# Block design

What is considered 'block' design, we reduce the number of alternatives in each choice situtation. For example, when testing many levels of the attributes for the alternatives, we can end up with a choice situation with many alternatives.
In the coffee example, imagine combining:
 * 2 different levels of price, size, serving time, sweetener
 * For the alternatives: latte, cappucino, long black

We would end up with a choice situation of $2 * 2 *2 * 2 *3=48$ alternatives to choose from!

By blocking, we split the choice set into subblocks. Then we present subblocks to different individuals, and we later join everything back together.

*NOTE: In design of experiments the term 'blocking' is also used to capture nuisance parameters, introducing variables in the design that are not interesting, but we need to control for in order to improve the estimation*

# Randomization

> "*Control what you can, randomize the rest*"

Imagine that men and women have different preferences for the products.
We we design an experiment, we cannot present all choice sets to all participants in the experiment.

So we create our choice situations.
To get data for our survey, Imagine that we go to a university to find our respondants. We have our choice cards, the startegy is to wait outside of the classrooms for students to go out and survey them for a few minutes. We divide our choice cards among the experimenters in our team.
One member of the team goes to a class that is majority of women, the other goes to a class that majority men students. When we get our data, there will be a correlation among the choice situations sample and gender, and the effect of the gender will not be captured, or what is worse, get biases, get the wrong conclusion.

The same can happen if we go to survey people in a particular area in the city, maybe is frequented by people of a certain age, maybe it is a working day so the income of the people there is not representative (people are at their offices, not on the streets, you will get unemployed, retired), etc.

Because it is generally *extremely difficult* for experimenters to eliminate bias using only their expert judgment, the use of randomization is key to fight against it.

There are many ways to randomize an experiment one of them is to gather the individuals, and assign them randomly to the choice sets units, usign for example a random number generator. This is **after the experiment has been designed, do not confuse it with a random design where a subset of the possible combinations is picked at random.**
